```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='muted')
%matplotlib inline
```

## 1. Load Dataset

```python
# Generate dataset if not present
import os
if not os.path.exists('../data/student_data.csv'):
    exec(open('../data/generate_dataset.py').read())

df = pd.read_csv('../data/student_data.csv')
print('Shape:', df.shape)
df.head()
```

## 2. Basic Information

```python
print('=== Data Types ===')
print(df.dtypes)
print('\n=== Missing Values ===')
print(df.isnull().sum())
print('\n=== Class Balance ===')
print(df['Result'].value_counts())
print('\n=== Grade Distribution ===')
print(df['FinalGrade'].value_counts())
```

## 3. Statistical Summary

```python
df.describe().round(2)
```

## 4. Distribution Plots

```python
num_cols = ['Attendance','StudyHours','MidTermMarks','AssignmentScore','PrevGPA','FinalMarks']
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
fig.suptitle('Feature Distributions', fontsize=14, fontweight='bold')
for ax, col in zip(axes.flat, num_cols):
    sns.histplot(df[col], kde=True, ax=ax, color='#2E74B5')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 5. Correlation Heatmap

```python
corr = df[num_cols].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, square=True)
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 6. Pass/Fail Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Feature vs. Result (Pass/Fail)', fontsize=13, fontweight='bold')
for ax, col in zip(axes, ['Attendance','MidTermMarks','StudyHours']):
    sns.boxplot(data=df, x='Result', y=col, ax=ax,
                palette={'Pass':'#27ae60','Fail':'#e74c3c'})
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## 7. Grade Distribution by Gender

```python
plt.figure(figsize=(10, 5))
order = ['O','A+','A','B+','B','F']
sns.countplot(data=df, x='FinalGrade', hue='Gender', order=order,
              palette='Set2')
plt.title('Grade Distribution by Gender', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## ✅ EDA Complete
Key findings:
- MidTermMarks and Attendance are strongly correlated with FinalMarks.
- Study hours show a positive relationship with passing.
- Dataset is well-balanced for Pass/Fail classification.
